```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./datasets/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.columns
```

```python
df.head()
```

```python
df['tenure'].dtype
```

```python
df['tenure'] = df['tenure'].fillna(df['tenure'].median())
```

```python
df['MonthlyCharges'].dtype
```

```python
df['MonthlyCharges'] = df['MonthlyCharges'].fillna(df['MonthlyCharges'].median())
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
df.isna().sum()
```

```python
conditions = [
    (df['tenure'] >= 0 ) & (df['tenure'] <= 12),
    (df['tenure'] >= 12) & (df['tenure'] <= 36),
    (df['tenure'] >= 37) & (df['tenure'] <= 60),
]

choices = ['New', 'Regular', 'Loyal']

df['tenure_group'] = np.select(conditions, choices, default='Champion')
```

```python
df[['tenure', 'tenure_group']]
```

```python
# 2. monthly_charge_segment
# MonthlyCharges < 30  → "Low"
# 30–70              → "Medium"
# > 70                 → "High"
conditions = [
    df['MonthlyCharges'] < 30,
    (df['MonthlyCharges'] >= 30) & (df['MonthlyCharges'] <= 70)
]

choices = ['Low', 'Medium']

df['monthly_charge_segment'] = np.select(conditions, choices, default='High')

df[['MonthlyCharges', 'monthly_charge_segment']]
```

```python
# 3. has_internet_service
# Convert InternetService column:
# "DSL" / "Fiber optic" → 1
# "No" → 0

df['has_internet_service'] = np.where(
    df['InternetService'] == 'No', 0, 1
)

df[['InternetService', 'has_internet_service']]
```

```python
# is_multi_line_user
# 1 if MultipleLines == "Yes"
# 0 otherwise

df['is_multi_line_user'] = np.where(
    df['MultipleLines'] == 'Yes', 1, 0
)

df[['MultipleLines', 'is_multi_line_user']]
```

```python
# 5. contract_type_code
# Map:
# Month-to-month → 0
# One year      → 1
# Two year      → 2

df['contract_type_code'] = df['Contract'].map({
    'Month-to-month': 0,
    'One year': 1,
    'Two year': 2
})
```

```python
df['Contract'].unique()
```

```python
df= df.drop(columns=['customerID', 'gender'], axis=1)
```

```python
df.to_csv('./datasets/preprocessedTeleComData.csv')
```
